# **1. 문자열**


```python
str1 = '오늘은 파이썬 첫날입니다'
str2 = "오늘은 파이썬 첫날입니다"
str3 = ''' 오늘은
날씨가 매우
덥습니다
하하하
'''

print(str1)
print(str2)
print(str3)

# '' 와 ""는 동일하고, '''  ''' 그 안에 쓰여진 그대로 출력된다.
```

# **2.문자열 인덱싱**

```python
str = "오늘은 파이썬 첫날입니다 날씨가 매우 덥습니다"

print(str[0])   # 오
print(str[1])   # 늘
print(str[10])  # 입
print(str[-1])  # 다(마지막 문자)
print(str[-2])  # 니(마지막 문자 앞 문)
```

# **3. 문자열 슬라이싱**

```python
str = "오늘은 파이썬 첫날입니다 날씨가 매우 덥습니다"

print(str[0:8])     # 인덱스 0부터 8이전까지 문자열
print(str[5:15])
print(str[16:])     # 인덱스 16부터 끝까지 문자열
print(str[:16])     # 인덱스 0부터 16 직전까지 문자열
```

# **4. 문자열 함수**

```python
str = "오늘은 파이썬 첫날입니다 날씨가 매우 덥습니다"

# len() : 인수로 입력된 문자열의 개수를 리턴, 파이썬 표준 함수
print(len(str))

# count() : 문자열에서 인수로 입력된 문자열의 출련 횟수를 리턴, 대소문자를 가린다
print(str.count('다'))
print(str.count('수'))

# find() : 인수로 입력된 문자열이 처음 나오는 위치를 인덱스로 리턴
print(str.find('은'))
print(str.find('z'))    # 없으면 -1

# index() : 인수로 입력된 문자열이 처음 나오는 위치를 인덱스로 리
print(str.index('은'))
print(str.index('z'))
```

```python
str = "오늘은 파이썬 첫날입니다 날씨가 매우 덥습니다 Apple"

# replace() : 문자열을 치환
print(str)
print(str.replace("덥습","춥습"))

# upper() : 대문자로 변환
# lower() : 소문자로 변환
print(str.lower())
print(str.upper())
```

```python
str = '           Hello             Python         '
print(str)

# lstrip() : 문자열 왼쪽 공백 제거
print(str.lstrip())

# rsripn() : 문자열 오른쪽 공백 제거
print(str.rstrip())

# strinp() : 문자열 양쪽 공백을 제거
print(str.strip())
```

```python
# join() : 문자열에 특정 문자열을 삽입
str = 'Python'
print("/".join(str))
```
