# **1.변수란?**

* 변수는 처리한 데이터를 저장시키는 기억 장소


```python
name = '홍길동'
age = 20

print(name)
print(age)

name = '이순신'
print(name)
```

# **2.변수의 이름을 작성하는 방법**
1. 값이 나타내는 핵심 단어
2. 반드시 문자로 시작해야하고, 예약어는 사용 할 수 없음
3. 영문자, 숫자, 특수문자(_)로 사용하며 대소문자를 구분

```python
num1 = 10
num2 = 3
print('%d / %d = %f' % (num1, num2, num1/num2))
```

# **3.변수의 데이터 타입**
* 데이터 타입(자료형) : 자료가 저장되는 형태를 의미

```python
name = '홍길동'
age = 20
weight = 50.5
isBool = False      # True

# type() : 변수의 데이터 타입을 반환해주는 함수

print(type(name))     # <class 'str'> : string이라는 class로 만들어졌다(문자형 타입)
print(type(age))      # 정수형 타입
print(type(weight))   # 실수형 타입
print(type(isBool))   # 논리형 타입
```

```python
# 필요없는 변수를 삭제하는 방법
del isBool

print(isBool)
```
