# **2. 셀의 종류**

* 코드 : 파이썬 코드를 실행할 수 있는 셀
* 텍스트 : 설명을 적거나 문구등을 적기 위한 텍스트 전용 셀(마크다운)

### 2-1. 셀의 삽입 방법
1. 상단의 '+ 코드', '+텍스트' 버튼을 눌러 삽입
2. 코드셀을 위에 삽입 : 컨트롤 + M,A
3. 코드셀을 아래에 삽입 : 컨트롤 + M,B

### 2-2. 셀을 변경하는 방법
1. 코드셀을 텍스트로 변경하는 방법 : 컨트롤 + M, M
2. 텍스트셀을 코드셀로 변경하는 방법 : 컨트롤 + M, Y

### 2-3. 셀을 삭제하는 방법
1. 우측의 휴지통 아이콘을 클릭하여 삭제
2. 컨트롤 + M,D




# **3. 셀의 실행**

1. 컨트롤 + 엔터 : 해당 셀을 실행하고 컨서는 해당 셀에 남음
2. 쉬프트 + 엔터 : 해당 셀을 실행하고 커서는 다음 셀로 넘어간다
