# **2.조건문**

## 2-1. if문

### **2-1-1. 비교 연산자**

* <, >, >=, <=, !(다르다), ==(같다)
* 결과가 True, False로 나온다

````python
print(10 > 5)
print(10 < 5)
print(10 >= 5)
print(10 <= 5)
print(10 != 5)
print(10 == 5)
````

### *2-2. if문*

```
  if 조건식(결과가 True 또는 False가 나오는 문장) : 
    조건식이 True인 경우 실행할 문
```

````python
age = int(input('나이를 입력하세요'))

if age > 19:
  print('성인입니다.')
  print('넷플릭스 성인 인증이 완료 되었습니다.')
print('입력한 나이 : ', age)
````

````python
age2 = int(input('나이를 입력하세요 '))

if age2 > 19 : print('성인입니다.')
print('입력한 나이 : ', age)
````

### **2-3. if-else 문**

```
  if 조건식(결과가 True 또는 False가 나오는 문장):
    조건식이 True인 경우 실행할 문장
    ....
  else :
    조건식이 False인 경우 실행할 문
```

````python
age = int(input('나이를 입력하세요'))

if age > 19 : 
  print('성인입니다.')
  print('넷플릭스 성인 인증 완료!.')
else :
  print('미성년입니다.')
  print('넷플릭스 성인 인증 실패!.')

print('입력한 나이 : ',age)
````

### **2-4. if-elif-else 문**

```
  if 조건식(결과가 True 또는 False가 나오는 문장):
    조건식이 True인 경우 실행할 문장
  elif 조건식2:
    조건식2이 True인 경우 실행할 문장
  elif 조건식3:
    조건식3이 True인 경우 실행할 문장
  else:
    조건식이 False이 경우 실행할 문
```

````python
age = int(input('나이를 입력하세요'))

if age > 19 : 
  print('성인입니다.')
elif age > 14:
  print('청소년입니다.')
elif age > 6:
  print('어린이입니다.')
else :
  print('입력이 잘못되었습니다.')

print('입력한 나이 : ', age)
````

#### **2-5. AND연산자**

* 양쪽 두 식의 값이 모두 True인 경우 결과가 True가 되는 연산자

|A|B|결과|
|----|----|----|
|True|True|True|
|True|False|True|
|False|True|False|
|False|False|False|

````python
# 로그인 프로그램 작성
# 아이디는 : admin, 비밀번호 : 1234

userid = input('아이디를 입력하세요')
userpw = input('비밀번호를 입력하세요')

if userid == 'admin' and userpw == '1234' :
  print('로그인 되어습니다.')
else :
  print('아이디 또는 비밀번호를 확인하세요.')
````

# **2-6. OR연산자**

* 양쪽 두 식의 값이 하나라도 True 인 경우 결과가 True가 되는 연산자

|A|B|결과|
|----|----|----|
|True|True|True|
|True|False|True|
|False|True|True|
|False|False|False|

````python
# 주민등록번호에 의한 성별 판별
# 001011-3068518
juminNo = input('주민등록번호 13자리를 - 없이 입력하세요') # 0010113068518

if juminNo[6] == '1' or juminNo[6] == '3':
  print('남자')
elif juminNo[6] == '2' or juminNo[6] == '4':
  print('여자')
else :
  print('잘못 입력했습니다.')
````

# **3. 대입연산자**

* =, +=, -=, *=, /=, %=
* 왼쪽에 저장된 값과 오른쪽의 값을 연산한 뒤 다시 왼쪽에 대입



````python
num = 10

num += 1    # num = num + 1
print(num)

num -= 1
print(num)

num *= 10
print(num)

num /= 10
print(num)

num %= 2
print(num)
````

# **4. while 문**

```
  while 조건식 :
    조건식이 True인 동안 반복할 문장
    ...
```

````python
i = 1
while i <= 5:
  print("hello Python")
  i += 1

print('현재 i의 값 : {}'.format(i))
````

````python
# 1~10까지의 합

i=1
sum=0

while i < 10:
  sum += i
  i += 1

print("1부터 10까지의 합 : {}".format(sum))
````
